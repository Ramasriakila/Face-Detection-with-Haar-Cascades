```python
face_cas=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect(im):
    face = im.copy()
    rect = face_cas.detectMultiScale(face)

    for (x, y, w, h) in rect:
        cv2.rectangle(face, (x, y), (x + w, y + h), (127, 0, 255), 10)
        cv2.imshow('Face Detection', face)

    return face
```

```python
res_img=detect(img)
plt.imshow(res_img[:,:,::-1])
plt.show()
```

```python
res_gls=detect(gls)
plt.imshow(res_gls[:,:,::-1])
plt.show()
```

```python
res_grp=detect(grp)
plt.imshow(res_grp[:,:,::-1])
plt.show()
```
